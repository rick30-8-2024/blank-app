```python
import streamlit as st
import requests
import time

# Define a function to simulate fetching JSON from a server
# @st.cache_data
def fetch_json_data():
    # Simulating a delay for fetching data
    time.sleep(3)
    
    # Simulated JSON data from the server
    return {
        "first_name": "John",
        "last_name": "Doe",
        "email": "johndoe@example.com",
        "phone": "+1234567890",
        "company_name": "Example Inc.",
        "company_type": "Technology",
        "user_role": "Developer"
    }

# Function to display loading spinner while fetching data
def loading_spinner():
    svg = '''
    <svg xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink" style="margin: auto; background: none; display: block; shape-rendering: auto;" width="100px" height="100px" viewBox="0 0 100 100" preserveAspectRatio="xMidYMid">
        <circle cx="50" cy="50" fill="none" stroke="#49d1e0" stroke-width="10" r="35" stroke-dasharray="164.93361431346415 56.97787143782138">
            <animateTransform attributeName="transform" type="rotate" repeatCount="indefinite" dur="1s" keyTimes="0;1" values="0 50 50;360 50 50"></animateTransform>
        </circle>
    </svg>
    '''
    return svg

# Main Streamlit app layout
st.title("JSON Fetcher with Expandable Component")
if st.button("Fetch JSON"):
    with st.spinner('Fetching data...'):
        with st.expander("Click to fetch and display JSON data"):
        
            placeholder = st.empty()  # Placeholder for loading spinner
            placeholder.markdown(loading_spinner(), unsafe_allow_html=True)  # Display SVG loader
            
            json_data = fetch_json_data()  # Fetch JSON data from server
            
            placeholder.empty()  # Remove loading spinner after fetching data
            st.success('Data fetched successfully!')
        
        # Display fetched JSON data
        st.json(json_data)
```

```python
#1


import streamlit as st

def main():
    st.set_page_config(layout="wide", page_title="AI-Researcher")

    if "button_clicked" not in st.session_state:
        st.session_state.search_button = False
    
    if "search_query" not in st.session_state:
        st.session_state.search_query = ""

    # LANDING PAGE
    if not st.session_state.button_clicked:
        st.header("AI-Researcher")

        # Create the text input below the markdown
        st.session_state.search_query = st.text_input("", placeholder="Search Query Here...", key="top_search")

        # When the Search button is clicked
        if st.button("Search"):
            st.session_state.button_clicked = True
            st.rerun()

    if st.session_state.search_button:
        st.markdown(
            """
            <style>
            .code-background {
                background-color: #1a1c24; 
                color: #fff;     
                padding-top: 3px;
                padding-left: 10px;
                border-radius: 10px;
                font-family: monospace;    
                font-size: 16px;
                white-space: pre-wrap;     
            }
            </style>
            """,
            unsafe_allow_html=True
        )
        st.subheader("Asked Question:")
        st.markdown(f'<div class="code-background"><pre>{st.session_state.search_query}</pre></div>', unsafe_allow_html=True)
        main_col, media_col = st.columns([3, 1])

        with main_col:
            # Asked Question section
            
            # Previous URLs section
            st.subheader("Source Urls:")
            urls_container = st.container()
            with urls_container:
                cols = st.columns(5)
                for i in range(5):
                    with cols[i]:
                        st.markdown(f"[URL {i+1}](https://example.com)")
                
                # +12 more indicator
                st.text("+12 more")
            
            # Answer section
            st.subheader("Answer to the Question")
            st.text_area("Answer:", height=200, key="answer")

        with media_col:
            # Images section
            st.subheader("Images")
            for i in range(2):
                st.image("https://via.placeholder.com/150", caption=f"Image {i+1}")
            
            # Videos section
            st.subheader("Videos")
            st.video("https://www.youtube.com/watch?v=dQw4w9WgXcQ")

        # Bottom search bar
        st.text_input("Search:", key="bottom_search")
        
        
        
        
        
        
        if st.button("Reset"):
            st.session_state.button_clicked = False
            st.rerun()













if __name__ == "__main__":
    main()














































```
